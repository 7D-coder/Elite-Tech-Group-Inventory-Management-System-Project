```python
import json
fb = open("Record.json","r")
txt = fb.read()
fb.close()
record = json.loads(txt)
#This is for importing  records of iems that we have for sell.

fb1 = open("sell_till_now.json","r")
txt1 = fb1.read()
fb1.close()
sells_record = json.loads(txt1)
#This is for importing records of how much items has been sold. 
```

```python
import datetime as da
print("       Welcome to Grocery shop      ")
print("---------------------------------------")
choice = int(input("Enter 1 to -> Purchase\nEnter 2 to ->Add a new item in Record\n"))
print("---------------------------------------")
if(choice==1):
    Bar_code = input("Enter the bar code of product ")
    qu = int(input("Enter the quantity "))
    if((record[Bar_code]["quantity"]-qu)<0):
        print("\nSorry the Product ",record[Bar_code]["name"]," is out of stock")
    else:
        l = record[Bar_code]["exp"]
        l1 = record[Bar_code]["manuf"]
        manu_date = da.date(l1[0],l1[1],l1[2])
        ex_date = da.date(l[0],l[1],l[2])
        if(ex_date < da.date.today()):
            print("\nSorry the Product",record[Bar_code]["name"],"is Expried")
        else:
            print("************************************")
            print("           Final Bill               ")
            print("Product: ",record[Bar_code]["name"])
            print("Price: ",record[Bar_code]["price"])
            print("Expiry date: ",ex_date)
            print("Manufacturing date: ",manu_date)
            print("Amount: ",(qu*record[Bar_code]["price"]))
            print("        Thanks for shopping         ")
            print("************************************")
            record[Bar_code]["quantity"] = record[Bar_code]["quantity"]- qu
            if(Bar_code in sells_record.keys()):
                sells_record[Bar_code]["quantity"] = sells_record[Bar_code]["quantity"] + qu 
                sells_record[Bar_code]["Amount"] = sells_record[Bar_code]["Amount"] + (qu*record[Bar_code]["price"])
            else:
                g = {}
                g["name"] = record[Bar_code]["name"]
                g["quantity"] = qu 
                g["Amount"] = qu*record[Bar_code]["price"] 
                sells_record[Bar_code] = g

elif(choice==2):
    bar_code = input("Enter the bar code of new product ")
    qun = int(input("Enter the Quantity ")) 
    if(bar_code in record.keys()):
        record[bar_code]["quantity"] = record[bar_code]["quantity"] + qun 
    else:
        g  = {} 
        n = input("Enter the name of product ")
        p = int(input("Enter the price of product "))
        print("Enter the expriey date")
        ex = []
        a = int(input("Enter the expriey year "))
        ex.append(a)
        b = int(input("Enter the expriey month "))
        ex.append(b)
        c = int(input("Enter the expriey date "))
        ex.append(c)
        print("Enter the Manufacturing date")
        ma = []
        a = int(input("Enter the Manufacturing year ")) 
        ma.append(a)
        b = int(input("Enter the Manufacturing month "))
        ma.append(b)
        c = int(input("Enter the Manufacturing date ")) 
        ma.append(c)
        g["name"] = n
        g["price"] = p 
        g["exp"] = ex 
        g["manuf"] = ma 
        g["quantity"] = qun 
        record[bar_code] = g
    print("\nThe product have been added")

    
js = json.dumps(record)
t = open("Record.json","w")
t.write(js)
t.close()
#Updated our records of items we have


js1 = json.dumps(sells_record)
t1 = open("sell_till_now.json","w")
t1.write(js1)
t1.close()
#Updated the records of sell 
        


```
